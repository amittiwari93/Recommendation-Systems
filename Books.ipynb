```python
import pandas as pd
import numpy as np
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
dataset=pd.read_csv('book.csv')
dataset
```

```python
dataset.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
dataset
```

```python
dataset.rename(columns={'User.ID':'user_id','Book.Title':'book_title','Book.Rating':'book_rating'},inplace=True)
```

```python
dataset
```

```python
dataset.isnull().sum()
```

```python
dataset['book_rating'].value_counts().sort_index()
```

```python
dataset['book_title'].value_counts()
```

```python
dataset.duplicated(subset=['user_id']).sum()
```

```python
user_data=dataset.pivot_table(index='user_id',columns='book_title',values='book_rating')
user_data
```

```python
user_data.fillna(0,inplace=True)
```

```python
user_data
```

```python
# calculating cosine similarity in users
sim=1-pairwise_distances(user_data.values,metric='cosine')
```

```python
sim
```

```python
# Storing data in dataframe format
user_sim=pd.DataFrame(sim)
user_sim
```

```python
# Set index and column names to user ID
user_sim.index=dataset.user_id.unique()
user_sim.columns=dataset.user_id.unique()
```

```python
user_sim
```

```python
# Changing diagonal values to 0 as model 
# would always find those value best suited.
np.fill_diagonal(sim,0)
user_sim
```

```python
user_sim.idxmax(axis=1)
```
